```python
from rdkit import Chem
```

```python
with open('./test/smiles.smi', 'r') as f:
    lines = f.readlines()
lines = [l.strip() for l in lines]
```

```python
def mol2smiles(smi):
    try:
        return Chem.MolToSmiles(Chem.MolFromSmiles(smi))
    except:
        return None
```

```python
print(len(lines))
smiles = [mol2smiles(s) for s in lines]
smiles = list(filter(None, smiles))
print(len(smiles))
```
